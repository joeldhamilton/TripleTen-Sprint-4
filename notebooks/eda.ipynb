This is a Jupyter Notebook that will be used to conduct Experimental Data Analysis for the Sprint 4 project for TripleTen.

I will be using a table called vehicles_us to analyze data from a car advertising company.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Import vehicles_us.csv
vehicles = pd.read_csv('~/TripleTen/TripleTen Sprint 4/vehicles_us.csv')

#Preliminary data exploration
display(vehicles.head())
vehicles.info()
```

There are some columns that need to be changed:

'price' should be float64 since it is a price and can go to decimals

'model_year' should be int64

'cylinders' should be int64

'is_4wd' should be bool, since it is a yes or no answer

```python
vehicles['price'] = vehicles['price'].astype('float64')
vehicles['model_year'] = vehicles['model_year'].astype('int64') 
vehicles['cylinders'] = vehicles['cylinders'].astype('int64')
vehicles['is_4wd'] = vehicles['is_4wd'].astype('bool')
```
